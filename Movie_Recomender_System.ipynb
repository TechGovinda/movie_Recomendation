**import ast**

**ast.literal_eval()**

 it is used to convert string of list to integer

```python
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter<=3:
     L.append(i['name'])
     counter+=1
    else:
      break

  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

Converting overview **column in list** for** concatinate with other column** which are **already in list format.**

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
'Sam Worthington'
'SamWorthington'
# remove space becz if we seperate every words then its meaning going to be different so remove space
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# now converting tags column (list into string)
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
def clean_text(text):

  text = re.sub('[^a-zA-Z]',' ',text)
  text = ' '.join([ps.stem(word) for word in text.split() if word not in set(stopwords.words('english'))])
  return text
```

```python
new_df['tags']=new_df['tags'].apply(clean_text)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()

print(vectors.shape)

print(len(vectors[0]))
print(new_df['tags'][0])
```

```python
cv.get_feature_names_out()
```

we going to calculate angle betn the vector with cosine similarity
 similarity value is betn (0-1)
 nearly 0 have less similarity nearly shows higgh similarity betn 2 words

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
sorted(similarity[0],reverse=True)[1:6]
```

we sorted the movie similarity based on similarity but also we want the** index to **fetch the **exact name **of the movie so we use **enumerator function** to have the **same index while sorting the similarity**. key=lambda x:x[1] it is used for **sorting with column 1** value not with index value.

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']== movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```

```python
recommend('Spectre')
```
